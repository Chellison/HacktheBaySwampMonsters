## Read In Data

Three levels of Hydrologic Unit Codes (HUCs) are used to find the propotion of land cover types in the vicinity of a station. We used ArcGIS to find the proportion of each land cover type in the HUCS upstream from a station.

### Create Models

All models use a basic set of inputs (latitude, longitude, month, and year). We add different combinations of HUC8 land cover percentage, HUC10 land cover percentage, HUC12 land cover percentage, and upstream land cover percentage. We use linear, random forest, and gradient boosting regression. We train on all year up until and including 2015. We then test on years 2016 to present.

```python
#We selected Nitrogen as the pollutant measure of interest

#select entries with Nitrogen
datadf = datadf.dropna(subset=['TOTAL NITROGEN'])

#get month and year
datadf['year'] = pd.DatetimeIndex(datadf['DateTime']).year
datadf['month'] = pd.DatetimeIndex(datadf['DateTime']).month

```

```python
#run some models!

#set train/test set
train = datadf[datadf['year']<=2015]
test = datadf[datadf['year']>2016]
```

```python
test
```

```python
#function for making linear models
def linearModel(depVar, train, test):
    X = train[depVar]
    Y = train['TOTAL NITROGEN']

    lr_tn = LinearRegression(normalize='True')
    lr_tn.fit(X,Y)
    score_train = lr_tn.score(X, Y)
    score_test = lr_tn.score(test[depVar], test['TOTAL NITROGEN'])
    return lr_tn, score_train, score_test
```

```python
#Linear Model- add more varaibles

#Basic, LLAMY variables
depVar = ['Latitude', 'Longitude','month','year']
LLAMY_res = linearModel(depVar, train, test)
print('train score:', LLAMY_res[1])
print('LLAMY test score: ', LLAMY_res[2])

#LLAMY variables and HUC12 landcover percentages
depVar = ['Latitude', 'Longitude','month','year']+HUC12_percent
LLAMY_HUC12_res = linearModel(depVar, train, test)
print('train score:', LLAMY_HUC12_res[1])
print('LLAMY HUC12 test score: ', LLAMY_HUC12_res[2])

#LLAMY variables, HUC12 and HUC10 landcover percentages
depVar = ['Latitude', 'Longitude','month','year']+HUC12_percent+HUC10_percent
LLAMY_HUC12_10_res = linearModel(depVar, train, test)
print('train score:', LLAMY_HUC12_10_res[1])
print('LLAMY HUC12 and HUC10 test score: ', LLAMY_HUC12_10_res[2])

#LLAMY variables, HUC12, HUC10, and HUC8 landcover percentages
depVar = ['Latitude', 'Longitude','month','year']+HUC12_percent+HUC10_percent+HUC8_percent
LLAMY_HUC12_HUC10_HUC8_res = linearModel(depVar, train, test)
print('train score:', LLAMY_HUC12_HUC10_HUC8_res[1])
print('LLAMY HUC12, HUC10, and HUC8 test score: ', LLAMY_HUC12_HUC10_HUC8_res[2])

#LLAMY variables, HUC12, HUC10, HUC8, and stream HUCs landcover percentages
depVar = ['Latitude', 'Longitude','month','year']+HUC12_percent+HUC10_percent+HUC8_percent+stream_percent
LLAMY_HUC12_HUC10_HUC8_stream_res = linearModel(depVar, train, test)
print('train score:', LLAMY_HUC12_HUC10_HUC8_stream_res[1])
print('LLAMY HUC12 test score: ', LLAMY_HUC12_HUC10_HUC8_stream_res[2])
```

```python
#More linear models- see what spatial extent is most useful (HUC12, HUC10, HUC8, or stream HUCS)

#LLAMY variables and HUC12 landcover percentages
depVar = ['Latitude', 'Longitude','month','year']+HUC12_percent
LLAMY_HUC12_res = linearModel(depVar, train, test)
print('train score:', LLAMY_HUC12_res[1])
print('LLAMY HUC12 test score: ', LLAMY_HUC12_res[2])

#just HUC12 landcover percentages
depVar = HUC12_percent
HUC12_res = linearModel(depVar, train, test)
print('train score:', HUC12_res[1])
print('HUC12 test score: ', HUC12_res[2])

#LLAMY variables and HUC10 landcover percentages
depVar = ['Latitude', 'Longitude','month','year']+HUC10_percent
LLAMY_HUC10_res = linearModel(depVar, train, test)
print('train score:', LLAMY_HUC10_res[1])
print('LLAMY HUC10 test score: ', LLAMY_HUC10_res[2])

#just HUC12 landcover percentages
depVar = HUC10_percent
HUC10_res = linearModel(depVar, train, test)
print('train score:', HUC10_res[1])
print('HUC10 test score: ', HUC10_res[2])

#LLAMY variables and HUC8 landcover percentages
depVar = ['Latitude', 'Longitude','month','year']+HUC8_percent
LLAMY_HUC8_res = linearModel(depVar, train, test)
print('train score:', LLAMY_HUC8_res[1])
print('LLAMY HUC18 test score: ', LLAMY_HUC8_res[2])

#just HUC8 landcover percentages
depVar = HUC8_percent
HUC8_res = linearModel(depVar, train, test)
print('train score:', HUC8_res[1])
print('HUC8 test score: ', HUC8_res[2])

#LLAMY variables and HUCs encompassing upstream path landcover percentages
depVar = ['Latitude', 'Longitude','month','year']+stream_percent
LLAMY_stream_res = linearModel(depVar, train, test)
print('train score:', LLAMY_stream_res[1])
print('LLAMY stream test score: ', LLAMY_stream_res[2])

#just HUCs encompassing upstream path landcover percentages
depVar = stream_percent
stream_res = linearModel(depVar, train, test)
print('train score:', stream_res[1])
print('stream test score: ', stream_res[2])
```

```python
#function for making linear models
def gbrModel(depVar, train, test):
    X = train[depVar]
    Y = train['TOTAL NITROGEN']

    lr_tn = GradientBoostingRegressor()
    lr_tn.fit(X,Y)
    score_train = lr_tn.score(X, Y)
    score_test = lr_tn.score(test[depVar], test['TOTAL NITROGEN'])
    return lr_tn, score_train, score_test
```

```python
#Gradient Boosting Model- add more varaibles

#Basic, LLAMY variables
depVar = ['Latitude', 'Longitude','month','year']
LLAMY_res = gbrModel(depVar, train, test)
print('train score:', LLAMY_res[1])
print('LLAMY test score: ', LLAMY_res[2])

#LLAMY variables and HUC12 landcover percentages
depVar = ['Latitude', 'Longitude','month','year']+HUC12_percent
LLAMY_HUC12_res = gbrModel(depVar, train, test)
print('train score:', LLAMY_HUC12_res[1])
print('LLAMY HUC12 test score: ', LLAMY_HUC12_res[2])

#LLAMY variables, HUC12 and HUC10 landcover percentages
depVar = ['Latitude', 'Longitude','month','year']+HUC12_percent+HUC10_percent
LLAMY_HUC12_10_res = gbrModel(depVar, train, test)
print('train score:', LLAMY_HUC12_10_res[1])
print('LLAMY HUC12 and HUC10 test score: ', LLAMY_HUC12_10_res[2])

#LLAMY variables, HUC12, HUC10, and HUC8 landcover percentages
depVar = ['Latitude', 'Longitude','month','year']+HUC12_percent+HUC10_percent+HUC8_percent
LLAMY_HUC12_HUC10_HUC8_res = gbrModel(depVar, train, test)
print('train score:', LLAMY_HUC12_HUC10_HUC8_res[1])
print('LLAMY HUC12, HUC10, and HUC8 test score: ', LLAMY_HUC12_HUC10_HUC8_res[2])

#LLAMY variables, HUC12, HUC10, HUC8, and stream HUCs landcover percentages
depVar = ['Latitude', 'Longitude','month','year']+HUC12_percent+HUC10_percent+HUC8_percent+stream_percent
LLAMY_HUC12_HUC10_HUC8_stream_res = gbrModel(depVar, train, test)
print('train score:', LLAMY_HUC12_HUC10_HUC8_stream_res[1])
print('LLAMY HUC12, HUC10, HUC8, and stream test score: ', LLAMY_HUC12_HUC10_HUC8_stream_res[2])
```

```python
#More linear models- see what spatial extent is most useful (HUC12, HUC10, HUC8, or stream HUCS)

#LLAMY variables and HUC12 landcover percentages
depVar = ['Latitude', 'Longitude','month','year']+HUC12_percent
LLAMY_HUC12_res = gbrModel(depVar, train, test)
print('train score:', LLAMY_HUC12_res[1])
print('LLAMY HUC12 test score: ', LLAMY_HUC12_res[2])

#just HUC12 landcover percentages
depVar = HUC12_percent
HUC12_res = gbrModel(depVar, train, test)
print('train score:', HUC12_res[1])
print('HUC12 test score: ', HUC12_res[2])

#LLAMY variables and HUC10 landcover percentages
depVar = ['Latitude', 'Longitude','month','year']+HUC10_percent
LLAMY_HUC10_res = gbrModel(depVar, train, test)
print('train score:', LLAMY_HUC10_res[1])
print('LLAMY HUC10 test score: ', LLAMY_HUC10_res[2])

#just HUC12 landcover percentages
depVar = HUC10_percent
HUC10_res = gbrModel(depVar, train, test)
print('train score:', HUC10_res[1])
print('HUC10 test score: ', HUC10_res[2])

#LLAMY variables and HUC8 landcover percentages
depVar = ['Latitude', 'Longitude','month','year']+HUC8_percent
LLAMY_HUC8_res = gbrModel(depVar, train, test)
print('train score:', LLAMY_HUC8_res[1])
print('LLAMY HUC12 test score: ', LLAMY_HUC8_res[2])

#just HUC8 landcover percentages
depVar = HUC8_percent
HUC8_res = gbrModel(depVar, train, test)
print('train score:', HUC8_res[1])
print('HUC8 test score: ', HUC8_res[2])

#LLAMY variables and HUCs encompassing upstream path landcover percentages
depVar = ['Latitude', 'Longitude','month','year']+stream_percent
LLAMY_stream_res = gbrModel(depVar, train, test)
print('train score:', LLAMY_stream_res[1])
print('LLAMY stream test score: ', LLAMY_stream_res[2])

#just HUCs encompassing upstream path landcover percentages
depVar = stream_percent
stream_res = gbrModel(depVar, train, test)
print('train score:', stream_res[1])
print('stream test score: ', stream_res[2])
```

```python
#gather all models of interest for feature importance plots
#gradient boosting regression with HUC (12, 10, 8, upstream)
#landcover precentages with and without LLAMY
all_gbr = [stream_res[0], LLAMY_stream_res[0],HUC12_res[0], LLAMY_HUC12_res[0],
           HUC10_res[0], LLAMY_HUC10_res[0],HUC8_res[0], LLAMY_HUC8_res[0]]
```

```python
#feature importance plots
titles = ['Upstream HUCs', 'LLAMY + Upstream HUCs',
         'HUC12s', 'LLAMY + HUC12s',
         'HUC10s', 'LLAMY + HUC10s',
         'HUC8s', 'LLAMY + HUC8s']

fig = plt.figure(figsize=(40,40))
for i in range(0,len(titles)):
    
    feature_importance = all_gbr[i].feature_importances_
    sorted_idx = np.argsort(feature_importance)
    pos = np.arange(sorted_idx.shape[0]) + .5
    plt.rcParams.update({'font.size': 16})
    plt.subplot(4,2,i+1)
    plt.barh(pos, feature_importance[sorted_idx], align='center')
    if (i % 2) == 0:
        names = landcover_types
    else:
        names = ['Latitude', 'Longitude','month','year']+landcover_types
    plt.yticks(pos, np.array(names)[sorted_idx])
    plt.title('Feature Importance (MDI) '+titles[i])

plt.savefig('varImportance_w_wo_LLAMY.jpg')
plt.show()
```

Between the Upstream HUCs and HUC12 models, pasture+hay and emergent herbaceous wetland seem to be of varying importance.
