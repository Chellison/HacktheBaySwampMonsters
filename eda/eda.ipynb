```python
df_no_problem = df[df.Problem.isna()]
```

```python
df_CBP = df_no_problem[df_no_problem.Database == 'CBP']
```

```python
df_CBP['DateTime'] = pd.to_datetime(df_CBP[['Date', 'Time']].apply(lambda x: ' '.join(x), axis=1))
```

```python
df_CBP = df_CBP.sort_values(['Station', 'Latitude', 'Longitude', 'HUC12_', 'HUCNAME_', 
         'FIPS_', 'STATE_', 'COUNTY_', 'DateTime'])
```

```python
pivoted_df_CBP = df_CBP[['Station', 'Latitude', 'Longitude', 'HUC12_', 'HUCNAME_', 
         'FIPS_', 'STATE_', 'COUNTY_', 'DateTime', 
         'MeasureValue', 'Parameter', 'ParameterName_CBP','SampleDepth']].\
        pivot_table(values=['MeasureValue', 'SampleDepth'], 
                    index=['Station', 'Latitude', 'Longitude', 'HUC12_', 'HUCNAME_', 
                           'FIPS_', 'STATE_', 'COUNTY_', 'DateTime'],
                    columns=['Parameter', 'ParameterName_CBP'])
```
